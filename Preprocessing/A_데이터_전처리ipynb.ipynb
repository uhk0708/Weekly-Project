```python
def drop_one_syllable(df):
    # 빈 문자열을 NA로 대체
    df = df.replace('', pd.NA)

    # 한 글자인 값을 NA로 대체하는 사용자 정의 함수
    def replace_one_syllable(value):
        if isinstance(value, str) and len(value) == 1:
            return pd.NA
        return value

    # 데이터프레임의 모든 값에 대해 사용자 정의 함수 적용
    df = df.apply(replace_one_syllable)

    # NA 값을 제거
    df_dropna = df.dropna()

    return df_dropna

data['리뷰'] = drop_one_syllable(data['리뷰'])
```

# N-GRAM 설정 [벡터]

```python
import pandas as pd
from konlpy.tag import Okt
from sklearn.feature_extraction.text import CountVectorizer
from collections import OrderedDict

okt = Okt()

# 리뷰를 형태소 분석하여 토큰화 후 문자열로 변환
def tokenize_review(review):
    tokens = okt.pos(review, norm=True, stem=True)
    return ' '.join([word for word, pos in tokens])

sample_data['토큰화된 리뷰'] = sample_data['리뷰'].apply(tokenize_review)

# n-gram 추출 설정 (2-gram과 3-gram 추출)
n_gram_range = (2, 3)
vectorizer = CountVectorizer(ngram_range=n_gram_range)

# 벡터화 및 n-gram 추출
X = vectorizer.fit_transform(sample_data['토큰화된 리뷰'])

# n-gram 단어 목록을 새로운 컬럼에 저장
def remove_duplicates(text):
    return ' '.join(OrderedDict.fromkeys(text.split()))

sample_data['n-gram'] = [' '.join(vectorizer.inverse_transform(x)[0]) for x in X]
sample_data['n-gram'] = sample_data['n-gram'].apply(remove_duplicates)

# 결과 출력
print(sample_data)
```

# TF-IDF 설정 [리뷰]

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from collections import defaultdict
# Corpus는 문자열 목록(문서)이어야 합니다.
corpus = sample_data['리뷰'].tolist()  # '리뷰' 열 전체를 목록으로 변환
# TF-IDF 벡터라이저 초기화 및 변환
vectorizer = TfidfVectorizer()
sp_matrix = vectorizer.fit_transform(corpus)

# 단어와 인덱스 매핑
word2id = defaultdict(lambda: 0)
for idx, feature in enumerate(vectorizer.get_feature_names_out()):
    word2id[feature] = idx

# TF-IDF 점수가 0.3 이상인 단어들만 추출하여 새로운 열에 저장
def extract_high_tfidf_words(sentence, tfidf_matrix, row_index, threshold=0.3):
    words = sentence.split()
    high_tfidf_words = [(word, tfidf_matrix[row_index, word2id[word]]) for word in words if tfidf_matrix[row_index, word2id[word]] >= threshold]
    return ' '.join([f"{word}:{score:.2f}" for word, score in high_tfidf_words])

sample_data['리뷰_tf'] = [extract_high_tfidf_words(sent, sp_matrix, i) for i, sent in enumerate(corpus)]

# 결과 출력
print(sample_data)
```

```python
from konlpy.tag import Okt
import pandas as pd

# 형태소 분류기 Okt() 활용, 명사만 추출
nlp = Okt()
# 각 리뷰에 대해 명사만 추출하고 문자열로 변환
sample_data['명사'] = sample_data['리뷰'].apply(lambda x: ' '.join(nlp.nouns(x)))
# 결과 출력
sample_data
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from collections import defaultdict
# Corpus는 문자열 목록(문서)이어야 합니다.
corpus = sample_data['명사'].tolist()  # '리뷰' 열 전체를 목록으로 변환
# TF-IDF 벡터라이저 초기화 및 변환
vectorizer = TfidfVectorizer()
sp_matrix = vectorizer.fit_transform(corpus)

# 단어와 인덱스 매핑
word2id = defaultdict(lambda: 0)
for idx, feature in enumerate(vectorizer.get_feature_names_out()):
    word2id[feature] = idx

# TF-IDF 점수가 0.3 이상인 단어들만 추출하여 새로운 열에 저장
def extract_high_tfidf_words(sentence, tfidf_matrix, row_index, threshold=0.3):
    words = sentence.split()
    high_tfidf_words = [(word, tfidf_matrix[row_index, word2id[word]]) for word in words if tfidf_matrix[row_index, word2id[word]] >= threshold]
    return ' '.join([f"{word}:{score:.2f}" for word, score in high_tfidf_words])

sample_data['리뷰_noun_tf'] = [extract_high_tfidf_words(sent, sp_matrix, i) for i, sent in enumerate(corpus)]
```
